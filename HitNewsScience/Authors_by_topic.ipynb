```python
author_top = df[['Author','Assigned Topic']]
merged = pd.merge(author_top,author_top,on = 'Assigned Topic')
merged = merged[merged['Author_x'] != merged['Author_y']]
merged[['Author_x','Author_y']].value_counts()
#merged.to_csv('authors_by_topics_no_dups.csv')
```

```python
out = []
for i in range(len(merged)):
    right = merged.iloc[i]['Author_x']
    left = merged.iloc[i]['Author_y']
    out.append(set([left,right]))
```

```python
merged['combined'] = pd.Series(out).map(lambda x: list(x)[0] + ' \t ' + list(x)[1])
temp = merged[['combined','Assigned Topic']].drop_duplicates()
temp.dropna(inplace = True)
temp['left'] = temp['combined'].map(lambda x: str(x).split('\t')[0])
temp['right'] = temp['combined'].map(lambda x: str(x).split('\t')[1])
```
